```python
import numpy as np, pandas as pd, seaborn as sns, sklearn, matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import matplotlib.gridspec as gridspec
```

```python
df = sns.load_dataset('iris')
df
```

```python
fig,ax = plt.subplots(3,3,figsize = (20,20))
g = gridspec.GridSpec(3, 3)
# violin plot
a = plt.subplot(g[0,:])
#sns.violinplot(data = df, ax = a, orient = 'h')
#sns.violinplot(df['petal_length'],ax = a)
sns.violinplot(df['sepal_width'],ax = a, orient = 'h')
a.set_title('Violin Plot')


# histogram plot
b = plt.subplot(g[1,0])
#sns.histplot(df['petal_length'],ax = b, kde = True)
#sns.histplot(data = df, x = 'petal_width', y = 'petal_width',ax = b)
sns.histplot(df['sepal_width'],ax = b, kde = True)
b.set_title('histogram')

# mean, median, mode
c = plt.subplot(g[1,1])
#sns.kdeplot(data = df, x = 'petal_width', y = 'sepal_width',ax = c )
sns.kdeplot(df['petal_width'],  ax = c)
median = df['petal_width'].median()
mean = df['petal_width'].mean()
std_dev = df['petal_width'].std()
for i in range(1, 4):
    plt.axvline(median + i*std_dev, color='gray')
    plt.axvline(median - i*std_dev, color='blue')
    plt.axvline(mean+i*std_dev, color = 'black')
    plt.axvline(mean - i*std_dev,color = 'green')
c.set_title('Density')


# scatter plot

d = plt.subplot(g[1,2])
sns.scatterplot(data = df, x = 'petal_width', y = 'sepal_width', hue = 'species', ax = d)
d.set_title('scatterplot')


#pie chart
e = plt.subplot(g[2,0])
df['species'].value_counts().plot.pie(autopct='%1.1f%%', explode=[0.1, 0, 0], ax=e)
e.set_title('piechart')

#countplot
f = plt.subplot(g[2,1])
#sns.countplot(data = df, x = 'species', ax = f)
sns.countplot(data = df, y = 'petal_length', ax = f)
f.set_title('countplot')

#swarnplot
g = plt.subplot(g[2,2])
sns.swarmplot(data = df, x = 'petal_width', y = 'sepal_width',hue = 'species')
```

```python
sns.catplot(data = df, x = 'petal_width', y = 'sepal_width', hue = 'species', kind = 'strip', col ='species' )
```
